```python
df["Snowfall"]= df["Snowfall"].fillna(df["Snowfall"].mode()[0])
df["PRCP"]= df["PRCP"].fillna(df["PRCP"].mode()[0])
df["SNF"]= df["SNF"].fillna(df["SNF"].mode()[0])
```

```python
df["MAX"] = df["MAX"].fillna(df["MAX"].mean())
df["MIN"] = df["MIN"].fillna(df["MIN"].mean())
df["MEA"] = df["MEA"].fillna(df["MEA"].mean())
```

```python
df[["Precip", "Snowfall", "PRCP", "SNF"]] = df[["Precip", "Snowfall", "PRCP", "SNF"]].apply(pd.to_numeric, errors="coerce")
```

```python
def clean_numeric(series):
    s=series.astype(str).str.strip()
    s=s.replace({"T":"0","M":np.nan})
    s=s.str.replace("<","",regex=False)
    s=s.str.replace(",","",regex=False)
    return pd.to_numeric(s, errors="coerce")
```

```python
df["Precip"] = clean_numeric(df["Precip"])
df["Snowfall"] = clean_numeric(df["Snowfall"])
df["PRCP"] = clean_numeric(df["PRCP"])
df["SNF"] = clean_numeric(df["SNF"])
```

```python
df.info()
```

```python
X=df[["MaxTemp","MinTemp","Precip","Snowfall","YR","MO","DA"]]
y=df["MeanTemp"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test=train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
sns.pairplot(df[["MaxTemp", "MinTemp", "Precip", "Snowfall", "MeanTemp"]])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression= LinearRegression()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer()
```

```python
X_train =imputer.fit_transform(X_train)
X_test=imputer.transform(X_test)
```

```python
regression.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
y_pred=regression.predict(X_test)
print("R^2:", r2_score(y_test,y_pred))
print("MSE:", mean_squared_error(y_test,y_pred))
print("MAE:", mean_absolute_error(y_test,y_pred))
plt.scatter(y_test, y_pred)
plt.show()
```

```python
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r') 
plt.xlabel("Real Value (y_test)")
plt.ylabel("Prediction (y_pred)")
plt.title("Real vs Prediction")
plt.show()
```

```python
print("Intercept:", regression.intercept_)
print("Coefficients:", regression.coef_)

for name, coef in zip(X.columns, regression.coef_):
    print(name, ":", coef)
```

```python
residuals = y_test - y_pred

plt.hist(residuals, bins=50)
plt.xlabel("Residual (Real - Prediction)")
plt.ylabel("Frequency")
plt.title("Distribution of Residuals")
plt.show()
```
